```python
import pandas as pd

# Define file path and company-year combinations
file_path = "/Users/allullaswarantej/Downloads/"
companies = ["Apple", "Microsoft", "Tesla"]
years = ["2022", "2023", "2024"]

# Dictionary to store financial data
financial_data = {}

# Load data for each company-year
for company in companies:
    for year in years:
        file_name = f"{company}_Financial_Report_{year}.csv"
        df = pd.read_csv(file_path + file_name)

        # Extract key financial metrics
        revenue = df.iloc[1, 1]  # Net Sales (Revenue)
        net_income = df.iloc[3, 1]  # Gross Margin (could be adjusted)
        operating_expenses = df.iloc[4, 1]  # Operating Expenses
        gross_margin_percentage = round((float(net_income) / float(revenue)) * 100, 2) if revenue else None

        # Store in dictionary
        financial_data[(company, year)] = {
            "Total Revenue": revenue,
            "Net Income": net_income,
            "Operating Expenses": operating_expenses,
            "Gross Margin %": gross_margin_percentage
        }

# Function to answer predefined queries
def chatbot_query(company, year, query_type):
    if (company, year) in financial_data:
        return financial_data[(company, year)].get(query_type, "Data not available")
    else:
        return "Company or Year data not found."

# Example Usage
print(chatbot_query("Apple", "2023", "Total Revenue"))  # Fetch Apple's 2023 revenue
print(chatbot_query("Tesla", "2024", "Net Income"))  # Fetch Tesla's 2024 net income
print(chatbot_query("Microsoft", "2022", "Gross Margin %"))  # Fetch Microsoft's 2022 gross margin
print(chatbot_query("Microsoft", "2023", "Total Revenue"))  # Fetch Microsoft's 2023 revenue
print(chatbot_query("Apple", "2022", "Gross Margin %"))  # Fetch Apple's 2022 gross margin
```

```python
import pandas as pd

# Define file path and company-year combinations
file_path = "/Users/allullaswarantej/Downloads/"
companies = ["Apple", "Microsoft", "Tesla"]
years = ["2022", "2023", "2024"]

# Load financial data into a dictionary
financial_data = {}

for company in companies:
    for year in years:
        file_name = f"{company}_Financial_Report_{year}.csv"
        file = file_path + file_name
        df = pd.read_csv(file)
        financial_data[(company, year)] = df

# Function to extract specific values from the dataset
def get_financial_value(company, year, key):
    df = financial_data.get((company, year))
    if df is not None:
        for row in df.itertuples():
            if key in str(row):
                return row[2]  # Adjust column index based on your dataset
    return "Data not found"

# Simple chatbot function
def simple_chatbot():
    print("üí∞ Welcome to the Financial Chatbot! Ask me questions:")
    while True:
        user_query = input("\nEnter your question (or type 'exit' to quit): ").strip().lower()

        if user_query == "exit":
            print("Goodbye! üëã")
            break
        elif "total revenue" in user_query:
            company = input("Enter company (Apple, Microsoft, Tesla): ").strip()
            year = input("Enter year (2022, 2023, 2024): ").strip()
            revenue = get_financial_value(company, year, "Net sales")
            print(f"üìä The total revenue of {company} in {year} is: {revenue}")
        elif "net income change" in user_query:
            company = input("Enter company (Apple, Microsoft, Tesla): ").strip()
            income_2023 = get_financial_value(company, "2023", "Net Income")
            income_2024 = get_financial_value(company, "2024", "Net Income")
            if isinstance(income_2023, (int, float)) and isinstance(income_2024, (int, float)):
                change = income_2024 - income_2023
                direction = "increased" if change > 0 else "decreased"
                print(f"üìà The net income of {company} has {direction} by {abs(change)} from 2023 to 2024.")
            else:
                print("‚ö†Ô∏è Could not retrieve net income data.")
        else:
            print("ü§ñ Sorry, I can only provide predefined financial information.")

# Run the chatbot
simple_chatbot()
```
